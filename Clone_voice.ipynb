```python
from TTS.api import TTS
from google.colab import files
from IPython.display import Audio
import torch

#Get device
device="cuda:0" if torch.cuda.is_available() else "cpu"
```

```python
print(TTS().list_models())
```

```python
tts = TTS(model_name="voice_conversion_models/multilingual/vctk/freevc24", progress_bar=False).to(device)
```

```python
#@title Insert your source audio and target voice
source_wav_path = "" # @param {"type":"string","placeholder":"Insert path here"}
target_wav_path = "" # @param {"type":"string","placeholder":"Insert path here"}


# Conversion
tts.voice_conversion_to_file(source_wav = source_wav_path, target_wav = target_wav_path, file_path = "output.wav")

files.download("output.wav")


# Auto Donwnload flag
auto_download_audio  = False # @param {type:"boolean"}
from google.colab import files
if auto_download_audio:
  files.download("output.wav")

Audio("output.wav", autoplay = True)
```
